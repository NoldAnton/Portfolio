```python
data['ceiling_height'].astype('float') # –µ—Å–ª–∏ –≤—ã—Å–æ—Ç–∞ –ø–æ—Ç–æ–ª–∫–∞ –≤—ã—à–µ 20 –º–µ—Ç—Ä–æ–≤, –ø–µ—Ä–µ–¥–≤–∏–≥–∞–µ–º –∑–∞–ø—è—Ç—É—é.
data['ceiling_height'] = np.where(data['ceiling_height'] >= 20, data['ceiling_height'] / 10, data['ceiling_height'] * 1)

data['is_apartment'] = data['is_apartment'].fillna(False) # –∑–∞–º–µ–Ω–∏–º –≤—Å–µ –ø—Ä–æ–ø—É—Å–∫–∏ –Ω–∞ –∑–Ω–∞—á–µ–Ω–∏–µ False

data['balcony'] = data['balcony'].fillna(0) # –∑–∞–º–µ–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –Ω–∞ 0

for d in data['rooms'].unique(): # –∑–∞–º–µ–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –∂–∏–ª–æ–π –ø–ª–æ—â–∞–¥–∏, –º–µ–¥–∏–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –∫–≤–∞—Ä—Ç–∏—Ä —Å —Ç–∞–∫–∏–º –∂–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∫–æ–º–Ω–∞—Ç
    data.loc[(data['rooms'] == d) & (data['living_area'].isna()), 'living_area'] = \
    data.loc[(data['rooms'] == d), 'living_area'].median()
    
for d in data['locality_name'].unique(): # —Ç–∞–∫–∂–µ –ø–æ—Å—Ç—É–ø–∞–µ–º —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —ç—Ç–∞–∂–µ–π, –±–µ—Ä–µ–º –º–µ–¥–∏–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—é –≥–æ—Ä–æ–¥–∞
    data.loc[(data['locality_name'] == d) & (data['floors_total'].isna()), 'floors_total'] = \
    data.loc[(data['locality_name'] == d), 'floors_total'].median()

for d in data['floors_total'].unique(): # –≤—ã—Å–æ—Ç–∞ –ø–æ—Ç–æ–ª–∫–∞, –±–µ—Ä–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —Ç–∏–ø–æ–≤—ã–º –¥–æ–º–∞–º —Å —Ç–µ–º–∂–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —ç—Ç–∞–∂–µ–π.
    data.loc[(data['floors_total'] == d) & (data['ceiling_height'].isna()), 'ceiling_height'] = \
    data.loc[(data['floors_total'] == d), 'ceiling_height'].mean()
    
for d in data['total_area'].unique(): # –∑–∞–º–µ–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏, —Å—Ä–µ–¥–Ω–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –∫–≤–∞—Ä—Ç–∏—Ä —Å —Ç–∞–∫–æ–π –∂–µ –∂–∏–ª–æ–π –ø–ª–æ—â–∞–¥—å—é
    data.loc[(data['total_area'] == d) & (data['kitchen_area'].isna()), 'kitchen_area'] = \
    data.loc[(data['total_area'] == d), 'kitchen_area'].mean()
```
